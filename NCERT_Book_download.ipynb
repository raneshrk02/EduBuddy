#Download and Format data

```python
def download_and_unzip(name, url):
    zip_path = f"{name}.zip"
    folder = f"{name}"
    print(f"Downloading {name}...")

    r = requests.get(url)
    with open(zip_path, "wb") as f:
        f.write(r.content)

    os.makedirs(folder, exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(folder)

    os.remove(zip_path)
    return folder
```

```python
def extract_text_from_folder(folder):
    texts = []
    for filename in sorted(os.listdir(folder)):
        if filename.endswith(".pdf"):
            chapter_path = os.path.join(folder, filename)
            with pdfplumber.open(chapter_path) as pdf:
                text = "\n".join(page.extract_text() or "" for page in pdf.pages)
                if text.strip():
                    texts.append({"text": text.strip()})
    return texts
```

```python
final_data = []
```

```python
for name, url in links.items():
    try:
        folder = download_and_unzip(name, url)
        texts = extract_text_from_folder(folder)
        final_data.extend(texts)
    except Exception as e:
        print(f"Error processing {name}: {e}")
        continue
```

#Write JSONL file

```python
with open("ncert_books_for_gpt2.jsonl", "w", encoding="utf-8") as f:
    for item in final_data:
        json.dump(item, f, ensure_ascii=False)
        f.write("\n")
```
